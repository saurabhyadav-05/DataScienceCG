```python
import os
import statistics
from collections import Counter
from google.colab import userdata
from groq import Groq

# API Key
os.environ["GROQ_API_KEY"] = userdata.get("GROQ_API_KEY")

# Groq Client
client = Groq(api_key=os.environ["GROQ_API_KEY"])

# Sales Data
sales_data = [12000, 15000, 10000, 18000, 13000]

# Python Analysis
mean_val = statistics.mean(sales_data)
median_val = statistics.median(sales_data)

counts = Counter(sales_data)
max_freq = max(counts.values())
modes = [k for k, v in counts.items() if v == max_freq and max_freq > 1]

highest = max(sales_data)
lowest = min(sales_data)

mode_result = modes if modes else "No unique mode"

# Prompt for business insights
prompt = f"""
You are a business data analyst.

Analyze the following sales data:
{sales_data}

Calculated values:
- Mean: {mean_val}
- Median: {median_val}
- Mode: {mode_result}
- Highest: {highest}
- Lowest: {lowest}

Now provide:
1. A short business analysis
2. Key insights
3. Practical recommendations

Write in simple professional language.
"""

# LLM Response
response = client.chat.completions.create(
    model="llama-3.3-70b-versatile",
    messages=[
        {"role": "system", "content": "You are a business data analyst."},
        {"role": "user", "content": prompt}
    ],
    temperature=0
)

# Final Output
print("Sales Data Analysis")
print("-------------------")
print("Mean:", mean_val)
print("Median:", median_val)
print("Mode:", mode_result)
print("Highest:", highest)
print("Lowest:", lowest)
print("\nBusiness Insights:\n")
print(response.choices[0].message.content)
```
