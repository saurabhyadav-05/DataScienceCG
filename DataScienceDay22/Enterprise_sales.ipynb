```python

# 📌 Problem Statement
# Enterprise sales teams often struggle to:
# - Analyze client requirements from RFPs (Requests for Proposals)
# - Match company offerings to client needs
# - Draft tailored proposals that stand out
# - Ensure consistency across documents
# 👉 So we build a multi-agent AI system that automates sales proposal preparation.

# 🤖 System Overview
# - 🔹 Agent 1: RFP Analyzer
# - Reads client’s request for proposal
# - Extracts key requirements, budget constraints, and evaluation criteria
# - 🔹 Agent 2: Solution Mapper
# - Matches company products/services to client needs
# - Suggests tailored value propositions and differentiators
# - 🔹 Agent 3: Proposal Writer
# - Generates a polished, client-specific proposal
# - Ensures alignment with brand tone and compliance standards

# 🔄 Workflow
# RFP Document → Requirement Analysis → Solution Mapping → Proposal Generation

```

```python
# ==============================
# 🚀 INSTALL
# ==============================
!pip install crewai litellm --upgrade

# ==============================
# 🔐 API KEY SETUP (FIXED)
# ==============================
import os

try:
    from google.colab import userdata
    os.environ["GROQ_API_KEY"] = userdata.get("GROQ_API_KEY")
except:
    os.environ["GROQ_API_KEY"] = "PASTE_YOUR_API_KEY_HERE"

if not os.environ.get("GROQ_API_KEY"):
    raise ValueError("❌ API KEY MISSING")

print("✅ API Key Loaded")

# ==============================
# 📦 IMPORTS
# ==============================
from crewai import Agent, Task, Crew

# ==============================
# 🤖 LLM
# ==============================
llm = "groq/llama-3.3-70b-versatile"

# ==============================
# 👨‍💼 AGENTS
# ==============================
rfp_analyzer = Agent(
    role="RFP Analyzer",
    goal="Extract requirements, budget, evaluation criteria",
    backstory="Expert in analyzing enterprise RFPs",
    llm=llm,
    verbose=True
)

solution_mapper = Agent(
    role="Solution Architect",
    goal="Map requirements to best solutions",
    backstory="Expert in enterprise systems",
    llm=llm,
    verbose=True
)

proposal_writer = Agent(
    role="Proposal Writer",
    goal="Generate a professional proposal",
    backstory="Expert in sales proposal writing",
    llm=llm,
    verbose=True
)

# ==============================
# 📄 INPUT
# ==============================
rfp_text = """
Client needs AI CRM system
Budget: $40,000

Requirements:
- Customer analytics
- Automation
- ERP integration

Evaluation:
- Cost
- Security
- Scalability
"""

# ==============================
# 📋 TASKS
# ==============================
task1 = Task(
    description=f"Analyze RFP:\n{rfp_text}",
    expected_output="Structured analysis",
    agent=rfp_analyzer
)

task2 = Task(
    description="Map solutions to requirements",
    expected_output="Solution mapping",
    agent=solution_mapper,
    context=[task1]
)

task3 = Task(
    description="Write full sales proposal",
    expected_output="Final proposal",
    agent=proposal_writer,
    context=[task2]
)

# ==============================
# 🤝 CREW (FIXED)
# ==============================
crew = Crew(
    agents=[rfp_analyzer, solution_mapper, proposal_writer],
    tasks=[task1, task2, task3],
    verbose=True,
    process="sequential"   # ✅ FIX
)

# ==============================
# 🚀 RUN
# ==============================
result = crew.kickoff()

print("\n====================")
print("FINAL PROPOSAL")
print("====================\n")
print(result)
```
