```python
#  Scenario: University Library Search
# Imagine you’re building a smart search system for a university library.
# - Problem: Students often type queries like “fast dog running” or “canine sprinting”, but
# the library catalog has books titled “The dog runs fast”. Traditional keyword search misses these connections.
# - Solution: Use embeddings to capture semantic meaning instead of exact words.


from sentence_transformers import SentenceTransformer

# Choose from popular models:
# "all-MiniLM-L6-v2" → 384 dims, lightweight, fast
# "all-mpnet-base-v2" → 768 dims, stronger semantic understanding
# "BAAI/bge-m3" → 1024 dims, multilingual + dense/sparse retrieval
# "sentence-transformers/e5-large-v2" → 1024 dims, top performer on MTEB

model_name = "all-MiniLM-L6-v2"   # swap this to test other models
model = SentenceTransformer(model_name)

texts = [
    "The dog runs fast",
    "A quick canine sprints",
    "Python is a programming language"
]

embeddings = model.encode(texts)

print(f"Model: {model_name}")
print(f"Embedding shape: {embeddings.shape}")   # e.g. (3, 384) for MiniLM
print(f"First vector sample: {embeddings[0][:10]}")  # show first 10 values
```

```python

# Scenario: Hospital Patient Record Search
# - Problem: Doctors and nurses often type queries like “heart attack” or “chest pain emergency”,
# but the hospital’s electronic health record (EHR) system stores diagnoses as “acute myocardial infarction”. Traditional keyword search fails because medical staff use everyday language while records use clinical terminology.
# - Solution: Use embeddings to capture semantic meaning. Embeddings understand that “heart attack” and “acute myocardial infarction” are related concepts. This allows the search system to return the right patient records even when the query wording doesn’t match the stored terminology.


# 🏥 Scenario: Hospital Patient Record Semantic Search
# Problem:
# Doctors may search using everyday terms like "heart attack" or "chest pain emergency",
# while the hospital database stores diagnoses using clinical terminology
# such as "acute myocardial infarction".
# Traditional keyword search fails because the words do not match exactly.

# Solution:
# Use sentence embeddings to capture semantic meaning.
# Embeddings convert text into numerical vectors that represent meaning,
# allowing the system to understand that:
# "heart attack" ≈ "acute myocardial infarction"

from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

# Choose embedding model
# "all-MiniLM-L6-v2" → fast and lightweight
# "all-mpnet-base-v2" → stronger semantic understanding
# "BAAI/bge-m3" → multilingual support
# "sentence-transformers/e5-large-v2" → top benchmark performance

model_name = "all-MiniLM-L6-v2"
model = SentenceTransformer(model_name)

# Hospital database records (stored medical terminology)
patient_records = [
    "acute myocardial infarction",
    "hypertension stage two",
    "type two diabetes mellitus",
    "chronic obstructive pulmonary disease",
    "asthma attack"
]

# Doctor's query (everyday language)
doctor_query = "heart attack emergency"

# Convert text to embeddings
record_embeddings = model.encode(patient_records)
query_embedding = model.encode([doctor_query])

# Compute similarity
similarities = cosine_similarity(query_embedding, record_embeddings)

# Find best matching medical record
best_match_index = similarities.argmax()
best_record = patient_records[best_match_index]

# Output results
print(f"Model Used: {model_name}")
print(f"Embedding Shape (records): {record_embeddings.shape}")
print(f"Query: {doctor_query}")
print(f"Best Matching Record: {best_record}")
print(f"Similarity Score: {similarities[0][best_match_index]:.4f}")

# Optional: show first 10 values of first embedding vector
print("\nSample Embedding Values (first record):")
print(record_embeddings[0][:10])
```

```python
# 🛒 Scenario: E-Commerce Product Search
# Problem:
# Customers may search using everyday phrases like "cheap running shoes"
# or "affordable sneakers for jogging",
# while the product catalog may store descriptions like
# "budget-friendly athletic footwear".
# Traditional keyword search may fail because the wording is different.

# Solution:
# Use sentence embeddings to capture semantic meaning.
# Embeddings convert text into numerical vectors that represent meaning,
# allowing the system to understand that:
# "cheap running shoes" ≈ "budget-friendly athletic footwear"
# 🛒 Scenario: E-Commerce Product Search (Using 4 Embedding Models)
# 🛒 Scenario: E-Commerce Product Search (4 models, error resolved)

from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

# Product catalog
product_catalog = [
    "budget-friendly athletic footwear",
    "premium leather formal shoes",
    "lightweight trail running sneakers",
    "casual cotton summer t-shirts",
    "water-resistant hiking backpack"
]

# Customer query
customer_query = "cheap running shoes"

# Correct model names
models = [
    "all-MiniLM-L6-v2",
    "all-mpnet-base-v2",
    "BAAI/bge-m3",
    "intfloat/e5-large-v2"   # fixed model id
]

for model_name in models:
    print("\n======================================")
    print(f"Running Model: {model_name}")

    model = SentenceTransformer(model_name)

    # E5 model needs query/passage prefixes
    if model_name == "intfloat/e5-large-v2":
        catalog_input = [f"passage: {item}" for item in product_catalog]
        query_input = [f"query: {customer_query}"]
    else:
        catalog_input = product_catalog
        query_input = [customer_query]

    # Generate embeddings
    catalog_embeddings = model.encode(catalog_input, normalize_embeddings=True)
    query_embedding = model.encode(query_input, normalize_embeddings=True)

    # Compute similarity
    similarities = cosine_similarity(query_embedding, catalog_embeddings)

    # Best match
    best_match_index = similarities.argmax()
    best_product = product_catalog[best_match_index]

    # Output
    print(f"Embedding Shape: {catalog_embeddings.shape}")
    print(f"Query: {customer_query}")
    print(f"Best Matching Product: {best_product}")
    print(f"Similarity Score: {similarities[0][best_match_index]:.4f}")
    print(f"Sample Embedding Values: {catalog_embeddings[0][:10]}")
```
