```python
# Scenario: Corporate Market Research & Strategy
# A company wants to explore launching a new product in a competitive market. The Manager Agent oversees the process and delegates tasks to specialized workers.

# 👩‍💼 Manager Agent
# - Receives the overall goal: “Evaluate feasibility of launching Product Y in Asia.”
# - Dynamically assigns tasks to worker agents depending on what’s needed.
# - Example: If budget is unclear → send to Finance Worker. If regulations are complex → send to Legal Worker.

# ================================
# 📊 Worker Agents
# ================================
# - Market Research Worker
# - Collects competitor data, customer preferences, and demand forecasts.
# - Reports: “High demand in Tier‑1 cities, moderate competition.”
# - Finance Worker
# - Analyzes budget, ROI, and pricing strategy.
# - Reports: “Estimated $10M investment, ROI in 2 years.”
# - Operations Worker
# - Evaluates supply chain, production capacity, and logistics.
# - Reports: “Factories can scale up, but shipping costs are high.”
# - Legal Worker
# - Reviews compliance, intellectual property, and regional regulations.
# - Reports: “Trademark available, but import laws require certification.”
# - HR Worker
# - Assesses staffing needs and training requirements.
# - Reports: “Need 50 new hires for customer support and sales.”
```

```python
# =========================================
# Multi-Agent Market Research System
# Using Groq API + Colab Secret Key
# =========================================

from groq import Groq
from google.colab import userdata

# ==============================
# LOAD API KEY FROM COLAB SECRET
# ==============================
GROQ_API_KEY = userdata.get("GROQ_API_KEY")

# Initialize client securely
client = Groq(api_key=GROQ_API_KEY)

# ==============================
# BASE AGENT CLASS
# ==============================
class Agent:
    def __init__(self, name, role):
        self.name = name
        self.role = role

    def run(self, task):
        prompt = f"""
        You are {self.name}, a {self.role}.
        Task: {task}
        Give a professional, structured business response.
        Avoid decorative symbols like *, emojis, or unnecessary formatting.
        """

        response = client.chat.completions.create(
            model="llama-3.3-70b-versatile",
            messages=[{"role": "user", "content": prompt}],
            temperature=0.7
        )

        return response.choices[0].message.content.strip()


# ==============================
# WORKER AGENTS
# ==============================
market_agent = Agent(
    "Market Research Agent",
    "expert in competitor analysis, demand forecasting, and customer insights"
)

finance_agent = Agent(
    "Finance Agent",
    "expert in investment planning, ROI, and pricing strategy"
)

operations_agent = Agent(
    "Operations Agent",
    "expert in supply chain, logistics, and manufacturing"
)

legal_agent = Agent(
    "Legal Agent",
    "expert in compliance, IP law, and regional regulations"
)

hr_agent = Agent(
    "HR Agent",
    "expert in hiring, staffing, and workforce planning"
)


# ==============================
# MANAGER AGENT
# ==============================
class ManagerAgent:
    def __init__(self):
        self.agents = {
            "market": market_agent,
            "finance": finance_agent,
            "operations": operations_agent,
            "legal": legal_agent,
            "hr": hr_agent
        }

    def decide_tasks(self, goal):
        return {
            "market": f"Analyze demand, competitors, and customer preferences for {goal}",
            "finance": f"Estimate budget, ROI, and pricing strategy for {goal}",
            "operations": f"Evaluate supply chain, production, and logistics for {goal}",
            "legal": f"Check compliance, certifications, and IP issues for {goal}",
            "hr": f"Estimate hiring and training needs for {goal}"
        }

    def run(self, goal):
        print("\n" + "="*50)
        print(f"GOAL: {goal}")
        print("="*50)

        tasks = self.decide_tasks(goal)
        results = {}

        for key, agent in self.agents.items():
            print("\n" + "-"*50)
            print(f"{agent.name} Processing")
            print("-"*50)

            result = agent.run(tasks[key])
            results[key] = result
            print(result)

        return results


# ==============================
# FINAL STRATEGY GENERATOR
# ==============================
def generate_final_strategy(results):
    combined = "\n\n".join(
        [f"{k.upper()} REPORT:\n{v}" for k, v in results.items()]
    )

    prompt = f"""
    You are a senior business strategist.

    Based on the following reports, provide:
    - Final Decision (Go or No-Go)
    - Key Risks
    - Strategy Plan

    Avoid decorative symbols and keep the output professional.

    {combined}
    """

    response = client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=[{"role": "user", "content": prompt}],
        temperature=0.7
    )

    return response.choices[0].message.content.strip()


# ==============================
# MAIN FUNCTION
# ==============================
def main():
    manager = ManagerAgent()

    goal = "Evaluate feasibility of launching Product Y in Asia"

    results = manager.run(goal)

    print("\n" + "="*50)
    print("FINAL STRATEGY")
    print("="*50)

    final_report = generate_final_strategy(results)
    print(final_report)


# ==============================
# RUN
# ==============================
if __name__ == "__main__":
    main()
```
