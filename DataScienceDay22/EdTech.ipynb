```python
# Scenario: Automated Learning Hub for an EdTech Company
# 🎓 Problem Statement
# An EdTech company wants to:
# - Keep track of the latest educational technology innovations
# - Automatically generate weekly blog posts for their website
# - Reduce time spent on manual research and content drafting
# 👉 So we build a multi-agent AI system using CrewAI to automate this workflow:
# - Trend Monitoring Agent → Continuously scans for new EdTech tools, teaching strategies, and AI-in-education updates
# - Content Summarization Agent → Condenses research into digestible insights tailored for educators and learners
# - Blog Writing Agent → Crafts engaging, SEO-friendly blog posts automatically
# - Publishing Agent → Schedules and posts content directly to the company’s website

!pip install crewai litellm langchain-community langchain-litellm

import os
from crewai import Agent, Task, Crew
from google.colab import userdata
from langchain_litellm import ChatLiteLLM

# -----------------------------
# 1. Set Groq API Key
# -----------------------------
os.environ["GROQ_API_KEY"] = userdata.get("GROQ_API_KEY")

# Remove OpenAI key if exists
if "OPENAI_API_KEY" in os.environ:
    del os.environ["OPENAI_API_KEY"]

# -----------------------------
# 2. Define LLM
# -----------------------------
llm = ChatLiteLLM(model="groq/llama-3.3-70b-versatile")

# -----------------------------
# 3. Define Agents
# -----------------------------

# 🔍 Trend Monitoring Agent
trend_agent = Agent(
    role="EdTech Trend Researcher",
    goal="Find latest trends in educational technology and AI in education",
    backstory="Expert in tracking global EdTech innovations and tools",
    llm=llm,
    verbose=True
)

# 🧠 Summarization Agent
summary_agent = Agent(
    role="Content Analyst",
    goal="Summarize research into simple and useful insights",
    backstory="Expert in simplifying complex educational trends",
    llm=llm,
    verbose=True
)

# ✍️ Blog Writing Agent
blog_agent = Agent(
    role="Blog Writer",
    goal="Write engaging and SEO-friendly blog posts for educators",
    backstory="Professional writer specializing in EdTech blogs",
    llm=llm,
    verbose=True
)

# 🚀 Publishing Agent
publish_agent = Agent(
    role="Publishing Manager",
    goal="Prepare and schedule blog content for publishing",
    backstory="Expert in handling CMS platforms and scheduling posts",
    llm=llm,
    verbose=True
)

# -----------------------------
# 4. Tasks
# -----------------------------

# Task 1: Research
task1 = Task(
    description="Find the latest 5 trends in EdTech, including AI tools, online learning platforms, and teaching strategies",
    expected_output="List of trends with short explanations",
    agent=trend_agent
)

# Task 2: Summarize
task2 = Task(
    description="Summarize the trends into clear key points for educators and students",
    expected_output="Bullet-point summary of insights",
    agent=summary_agent,
    context=[task1]
)

# Task 3: Blog Writing
task3 = Task(
    description="Write a well-structured, engaging, and SEO-friendly blog post based on the summarized insights",
    expected_output="Complete blog with title, headings, and conclusion",
    agent=blog_agent,
    context=[task2]
)

# Task 4: Publishing
task4 = Task(
    description="Prepare the blog for publishing, including formatting and scheduling details",
    expected_output="Final formatted blog ready for publishing",
    agent=publish_agent,
    context=[task3]
)

# -----------------------------
# 5. Crew
# -----------------------------
crew = Crew(
    agents=[trend_agent, summary_agent, blog_agent, publish_agent],
    tasks=[task1, task2, task3, task4],
    verbose=True
)

# -----------------------------
# 6. Run
# -----------------------------
result = crew.kickoff()

print("\n✅ FINAL OUTPUT:\n")
print(result)

```
