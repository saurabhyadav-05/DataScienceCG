```python
from groq import Groq
from google.colab import userdata

# Load API key from Colab Secret
GROQ_API_KEY = userdata.get("GROQ_API_KEY")

client = Groq(api_key=GROQ_API_KEY)

MODEL_NAME = "llama-3.3-70b-versatile"


# Base Agent Class
class Agent:
    def __init__(self, name, role):
        self.name = name
        self.role = role

    def respond(self, message):
        prompt = f"""
        You are {self.name}, a {self.role}.

        Situation:
        {message}

        Provide a clear and professional action plan.
        Keep the response natural like a human explanation.
        Avoid symbols, emojis, or decorative formatting.
        """

        response = client.chat.completions.create(
            model=MODEL_NAME,
            messages=[{"role": "user", "content": prompt}],
            temperature=0.5
        )

        return response.choices[0].message.content.strip()


# Crisis Coordinator (Broadcast)
class CrisisCoordinator:
    def __init__(self, message):
        self.message = message

    def broadcast(self):
        print("\nCrisis Alert:")
        print(self.message)
        return self.message


# Decision Agent (Final Coordinator)
class DecisionAgent:
    def generate_final_plan(self, responses):
        combined = "\n\n".join(
            [f"{name} Response:\n{resp}" for name, resp in responses.items()]
        )

        prompt = f"""
        You are a senior crisis management coordinator.

        Based on the following responses, create a final unified crisis response plan.

        Keep it simple, structured, and human-like.
        Do not use symbols or decorative formatting.

        {combined}
        """

        response = client.chat.completions.create(
            model=MODEL_NAME,
            messages=[{"role": "user", "content": prompt}],
            temperature=0.5
        )

        return response.choices[0].message.content.strip()


# Create Worker Agents
it_agent = Agent(
    "IT Security Agent",
    "expert in cybersecurity, threat mitigation, and system protection"
)

comm_agent = Agent(
    "Communications Agent",
    "expert in public relations and stakeholder communication"
)

finance_agent = Agent(
    "Finance Agent",
    "expert in financial planning and risk assessment"
)

legal_agent = Agent(
    "Legal Agent",
    "expert in compliance and corporate law"
)

hr_agent = Agent(
    "HR Agent",
    "expert in employee management and internal communication"
)


# Main Execution
def main():
    crisis_message = "Data breach detected in customer database. Immediate response required."

    coordinator = CrisisCoordinator(crisis_message)
    message = coordinator.broadcast()

    agents = {
        "IT Security": it_agent,
        "Communications": comm_agent,
        "Finance": finance_agent,
        "Legal": legal_agent,
        "HR": hr_agent
    }

    responses = {}

    for name, agent in agents.items():
        print(f"\n{name} Agent Response:")
        response = agent.respond(message)
        responses[name] = response
        print(response)

    decision_agent = DecisionAgent()

    print("\nFinal Crisis Response Plan:")
    final_plan = decision_agent.generate_final_plan(responses)
    print(final_plan)


if __name__ == "__main__":
    main()
```
