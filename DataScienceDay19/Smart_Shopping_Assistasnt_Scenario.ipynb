```python
# Scenario: Smart Shopping Assistant
# You say to your students:
# “Imagine you are planning a party and you tell your assistant:
# ‘Find me the best chocolate cake under ₹500.’
# Now instead of just replying with text, your assistant actually thinks, checks
# different bakery websites, compares prices, and then gives you the final decision.”
```

```python
# Define Tools
def search_products(product_name):
    return [
        {"name": "Chocolate Cake - Bakery A", "price": 450, "rating": 4.5},
        {"name": "Chocolate Cake - Bakery B", "price": 550, "rating": 4.7},
        {"name": "Chocolate Cake - Bakery C", "price": 400, "rating": 4.2}
    ]

def check_availability(product):
    return {"product": product["name"], "available": True}

def filter_budget(products, max_price):
    return [p for p in products if p["price"] <= max_price]
```

```python
# Agent Brain (Planning)
def agent_brain(user_query):
    plan = []

    if "cake" in user_query.lower():
        plan.append("search_products")
        plan.append("filter_budget")
        plan.append("check_availability")
        plan.append("choose_best")

    return plan
```

```python
# Execution Engine
def execute_plan(plan):
    context = {}

    for step in plan:

        if step == "search_products":
            print("\n[Action] Searching Products...")
            context["products"] = search_products("chocolate cake")
            print(context["products"])

        elif step == "filter_budget":
            print("\n[Action] Filtering by Budget...")
            context["filtered"] = filter_budget(context["products"], 500)
            print(context["filtered"])

        elif step == "check_availability":
            print("\n[Action] Checking Availability...")
            available = []
            for p in context["filtered"]:
                if check_availability(p)["available"]:
                    available.append(p)
            context["available"] = available
            print(context["available"])

        elif step == "choose_best":
            print("\n[Action] Choosing Best Product...")
            best = max(context["available"], key=lambda x: x["rating"])
            context["best_product"] = best
            print(best)

    return context
```

```python
# Full Agent
def shopping_agent(user_query):
    print("User Query:", user_query)

    plan = agent_brain(user_query)
    print("\n[Plan]", plan)

    result = execute_plan(plan)

    final_output = f"""
Best Product: {result['best_product']['name']}
Price: ₹{result['best_product']['price']}
Rating: {result['best_product']['rating']}
Recommendation: Best rated cake within your budget.
"""
    return final_output
```

```python
# Run the Agent
response = shopping_agent("Find best chocolate cake under 500")
print("\nFinal Answer:\n", response)
```
