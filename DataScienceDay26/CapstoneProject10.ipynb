```python
# 🎯 Objective:
# Build a multi-step intelligent advisor system using agent workflows.
# 💼 Problem Statement
# Students need guidance on career paths based on their skills.
# Your system should:
# Analyze student profile
# Suggest career roles
# Generate learning roadmap
# 👥 Agents to Build
# Profile Analyzer Agent
# Understands student background
# Career Recommendation Agent
# Suggests suitable roles
# Skill Gap Agent
# Identifies missing skills
# Learning Path Agent
# Creates step-by-step roadmap
# ⚙️ Tasks to Implement
# Sequential workflow (LangGraph preferred)
# State passing between agents
# Conditional logic (beginner vs intermediate)
# 💡 Sample Input
# “I know Python basics and statistics, and I want to become a Data Scientist”

# 🧪 Expected Output
# Suggested roles (e.g., Data Scientist, Analyst)
# Skill gaps (ML, SQL, etc.)
# Learning roadmap (step-by-step)
# 🧰 Tools (Choose Any)
# LangGraph (recommended)
# CrewAI
# AutoGen
# Python logic-based workflow
# Microsoft Copilot Studio
# 📦 Deliverables
# Workflow implementation
# Output for at least 2 student profiles
# Diagram of agent flow


# 🎓 AI CAREER ADVISOR MULTI-AGENT SYSTEM

!pip install groq

from groq import Groq
from google.colab import userdata

# 🔐 LOAD API KEY
client = Groq(api_key=userdata.get("GROQ_API_KEY"))

# 🧠 AGENT 1: PROFILE ANALYZER
def profile_analyzer(query):

    prompt = f"""
    Analyze the student profile.

    Extract:
    - Skills
    - Experience Level (Beginner / Intermediate / Advanced)
    - Career Goal

    Return in simple text.

    Query: {query}
    """

    res = client.chat.completions.create(
        model="llama-3.1-8b-instant",
        messages=[{"role": "user", "content": prompt}]
    )

    return res.choices[0].message.content


# 💼 AGENT 2: CAREER RECOMMENDATION
def career_agent(profile):

    prompt = f"""
    Based on the student profile below, suggest 2-3 career roles.

    Profile:
    {profile}

    Return only roles.
    """

    res = client.chat.completions.create(
        model="llama-3.1-8b-instant",
        messages=[{"role": "user", "content": prompt}]
    )

    return res.choices[0].message.content


# 📉 AGENT 3: SKILL GAP ANALYZER
def skill_gap_agent(profile, careers):

    prompt = f"""
    Identify missing skills required for these careers.

    Profile:
    {profile}

    Careers:
    {careers}

    Return bullet points.
    """

    res = client.chat.completions.create(
        model="llama-3.1-8b-instant",
        messages=[{"role": "user", "content": prompt}]
    )

    return res.choices[0].message.content


# 📚 AGENT 4: LEARNING PATH GENERATOR
def learning_path_agent(profile, gaps):

    prompt = f"""
    Create a step-by-step learning roadmap.

    Consider:
    - If beginner → start from basics
    - If intermediate → skip basics

    Profile:
    {profile}

    Skill Gaps:
    {gaps}

    Give structured steps (1,2,3...).
    """

    res = client.chat.completions.create(
        model="llama-3.1-8b-instant",
        messages=[{"role": "user", "content": prompt}]
    )

    return res.choices[0].message.content


# 🎯 FINAL RESPONSE FORMATTER
def final_response(query, profile, careers, gaps, roadmap):

    return f"""

🎓 AI CAREER ADVISOR REPORT

🧠 Input:
{query}

📊 Profile Analysis:
{profile}

💼 Recommended Careers:
{careers}

📉 Skill Gaps:
{gaps}

📚 Learning Roadmap:
{roadmap}

"""


# 🔁 MAIN SYSTEM (LOOP WITH EXIT)
def run_system():

    print("🎓 AI Career Advisor Started")
    print("Type 'exit' to stop\n")

    while True:

        query = input("Enter your profile: ")

        if query.lower() == "exit":
            print("👋 Exiting system...")
            break

        print("\n🔍 Analyzing profile...")
        profile = profile_analyzer(query)

        print("💼 Recommending careers...")
        careers = career_agent(profile)

        print("📉 Finding skill gaps...")
        gaps = skill_gap_agent(profile, careers)

        print("📚 Generating roadmap...")
        roadmap = learning_path_agent(profile, gaps)

        result = final_response(query, profile, careers, gaps, roadmap)
        print(result)


# ▶️ RUN
run_system()
```
